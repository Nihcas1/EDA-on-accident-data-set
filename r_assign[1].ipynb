```python
from bs4 import BeautifulSoup as bs
import requests 

url="https://www.news18.com/news/auto/road-safety-this-country-tops-list-of-worlds-worst-drivers-india-ranks-at-4th-place-7141687.html"
headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}

page=requests.get(url, headers=headers)
soup=bs(page.content,'html.parser')
content=soup.find('div',class_="jsx-2889061695").text
content
```

```python
import pandas as pd
data=pd.DataFrame({"article":content},index=[1])
data
```

```python
import nltk                             
from nltk.corpus import twitter_samples   
import matplotlib.pyplot as plt           
import random  
import pandas as pd
```

```python
#Removing punctuation
import string
string.punctuation
#Defining Function to remove punctuation
def remove_punctuation(data):
    punctuationfree="".join([i for i in data if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
data['punctuation']= data['article'].apply(lambda x:remove_punctuation(x))
data
```

```python
#Lowering the tweets
data['lower_text']= data['punctuation'].apply(lambda x: x.lower())
data
```

```python
#Tokenization

from nltk.tokenize import TweetTokenizer as tt
#applying function to the column
tokenizer = tt()      # instantiate the tokenizer class

data['tokenized']=data['lower_text'].apply(lambda x: tokenizer.tokenize(x))
data
#Removing stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopword = stopwords.words('english')

#Defining the function to remove stopwords from tokenized text
def remove_stopwords(data):
    output= [i for i in data if i not in stopword]
    return output
#applying the function
data['no_stopwords']= data['tokenized'].apply(lambda x:remove_stopwords(x))
data

#Stemming 
from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()
#Defining a function for stemming
def stemming(data):
    stem_tweet = [porter_stemmer.stem(word) for word in data]
    return stem_tweet
data['stemmed_tweets']=data['no_stopwords'].apply(lambda x: stemming(x))
data
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
wordnet_lemmatizer = WordNetLemmatizer()
#Defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in data]
    return lemm_text
data['lemmatized_tweets']=data['stemmed_tweets'].apply(lambda x:lemmatizer(x))
data
```
